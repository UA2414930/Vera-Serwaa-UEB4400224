```python
name = input("Enter your Name")

age = int(input("Enter your Age"))

if age >= 18:

    age_status = "Adult"

else:

    age_status = "Minor"

gpa = float(input("Enter your GPA"))

if gpa > 4.0:

    print("Invalid GPA")

elif gpa >= 3.5:

    deans_list = "Yes"

else:

    deans_list = "No"

courses = input("Enter your courses separated by commas: ")

course_list = courses.split(",")

student_profile = {}

student_profile["name"] = name

student_profile["age"] = age

student_profile["gpa"] = gpa

student_profile["courses"] = course_list

student_profile["num_courses"] = len(course_list)

print("\n======= STUDENT PROFILE =======")

print(f"Name: {name}")

print(f"Age: {age}")

print(f"Age Status: {age_status}")

print(f"GPA: {gpa}")

print(f"Dean's List: {deans_list}")

print(f"Courses: {course_list}")

print(f"Number of Courses: {len(course_list)}")

print("\n===== TYPE CONVERSION =====")

print(f"Name Type: {type(name)}")

print(f"Age Type: {type(age)}")

print(f"GPA Type: {type(gpa)}")

print(f"Courses Type: {type(course_list)}")

print(f"Student Profile Type: {type(student_profile)}")

print("==============================")
```
